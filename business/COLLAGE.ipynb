```python



import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("Auto Sales data.csv")
data
```

```python
# data.dtypes
# data.describe()
# data.info()
```

### check for NULL & DUPLICATE VALUE

```python
data.isnull().sum()
data.drop_duplicates(inplace=True)
```

```python
# Change data type
data['ORDERDATE'] = pd.to_datetime(data['ORDERDATE'])
```

```python
data
```

# Data cleaning & Missing value Handling

```python
# Drop column
data = data.drop(['ADDRESSLINE1','PHONE','CONTACTLASTNAME','CONTACTFIRSTNAME','CITY','POSTALCODE','DEALSIZE'], axis=1)
```

```python
data
```

```python
#sort
data_sort = data.sort_values(by='QUANTITYORDERED',ascending=False)
data_sort
```

```python
#top most 5 sales car
top_5_sales = data_sort.head(10)
top_5_sales.sort_values(by='ORDERNUMBER',ascending=True)
```

```python
#Visualization
sns.barplot(data=top_5_sales,x='ORDERNUMBER',y='QUANTITYORDERED',color='green')
plt.title("Top 5 Most Ordered car")
plt.xlabel('ORDERNUMBER')
plt.ylabel('ORDER')
```

```python
# Most prices cars
high_price = data_sort[data_sort.PRICEEACH > 100]
high_price
```

```python
# check 'car price' which nave greater then the 'MSRP'
OverPrice = data_sort[data_sort.PRICEEACH >= data_sort.MSRP]
OverPrice
```

```python
# Drop column
data_sort = data_sort.drop(['CUSTOMERNAME'], axis=1)
data_sort
```

```python
data_sort['MONTH'] = data_sort['ORDERDATE'].dt.month
data_sort
```

```python
# change data type
data_sort['QUANTITYORDERED'] = pd.to_numeric(data_sort['QUANTITYORDERED'])
data_sort
```

```python
#Monthly Ordered
monthly_order = data_sort.groupby('MONTH')['QUANTITYORDERED'].sum().reset_index()


top_5_count = monthly_order.sort_values(by='QUANTITYORDERED', ascending=False).head()
top_5_count
```

# BAR PLOT

```python
sns.barplot(data=top_5_count, x='MONTH', y='QUANTITYORDERED', color='pink')
plt.title('Succesfull car Sales Company')
plt.xlabel("MONTH")
plt.ylabel("Car Quantity")
```

# LINE PLOT

### Time Series Analytics

```python
import matplotlib.dates as mdates

data_sort['month_year'] = data_sort['ORDERDATE'].dt.to_period('M')
monthly_sales_data = data_sort.groupby('month_year')['QUANTITYORDERED'].sum()

#type(monthly_sales.index)
monthly_sales_data.index = monthly_sales_data.index.to_timestamp()

plt.figure(figsize=(8,6))
plt.plot(monthly_sales_data.index, monthly_sales_data.values)

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=6))
plt.xlabel('Month-Year')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from scipy import stats

z_score = np.abs((data[['PRICEEACH','MONTH']] -data[['PRICEEACH','MONTH']].mean())/data[['PRICEEACH','MONTH']].std())

outlier_threshold = 3
outlier = data[( z_score['PRICEEACH'] > outlier_threshold ) | ( z_score['MONTH'] > outlier_threshold )]

# outlier = z_score > threshold

print(outlier)
```
